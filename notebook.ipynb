```python
import matplotlib.pyplot as plt
import pandas as pd
import csv
from statistics import mean
```

# Amazon top sales

This notebook is focus to make an analyze to see top sales amazon for posible cases of dropshiping


```python
file_path = './data_sets/amazon.csv'

file = open(file_path, 'r')
reader = csv.reader(file, delimiter=',')
headers = next(reader)
data = []
for row in reader:
    iterable = zip(headers, row)
    sales_dict = {key: value for key, value in iterable}
    data.append(sales_dict)

labels = list(set([item['category'] for item in data]))
values = {}

for item in data:
    for category in item['category'].split('|'):
        values[category] = float(item['actual_price'][1:].replace(',', ''))

min_value = min(values.values())
max_value = max(values.values())
mean_value = mean(values.values())

highest_values = {key: value for key, value in values.items() if value > mean_value and value}

fig, ax = plt.subplots(figsize=(12, 5))
ax.bar(highest_values.keys(), highest_values.values())
plt.xticks(rotation=90)
plt.show()
```
